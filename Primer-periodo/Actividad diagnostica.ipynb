## Definimos nuestro dado

Para esta parte definiremos nuestro dado con una lista, de igual manera importamos la paqueteria necesaria

```python
import matplotlib.pyplot as plt
from collections import Counter
import random

dado = [1,2,3,4,5,6]
resultados = []
```

Posteriormente procedemos a simular el comportamiento de un dado cuando es lanzado 1000 veces, para esto usaremos un ciclo *for* y la funcion de *random* para simular el lanzamiento del dado y gegistramos los resultados en una lista

```python
for i in range(1000):
    resultados.append(random.choice(dado))
freq = Counter(resultados)
freq = (dict(sorted(freq.items())))

print(freq)
```

Utilizamos la paqueteria de *collections* para obtener la frecuencia de los elementos de la lista 

## Generamos nuestra grafica

Hacemos dos listas con los valores de *x* y de *y*, obtenidos de nuestro conjunto de pruebas 

```python
x_valores = []
y_valores = []

for key , value in freq.items():
    x_valores.append(key)
    y_valores.append(value)
    
print(x_valores)
print(y_valores)
```

Procedemos a crear nuestra grafica:

```python
plt.plot(x_valores, y_valores)
plt.xlabel("Cara del dado")
plt.ylabel("Resultados")
plt.title('Pruebas con el dado')
plt.show()
```

Calculamos el promedio de nuestros resultados

```python
media = 0
for x in y_valores:
    media += x
print('Media de nuestros valores: ', media/6)
```
